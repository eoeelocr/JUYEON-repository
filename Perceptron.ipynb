```python
# 1. 라이브러리 임포트 및 GPU 설정
import torch
import torch.nn as nn

# GPU 사용 설정
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# Random 함수 사용을 위한 seed value 설정
torch.manual_seed(2022)
if device == 'cuda':
  torch.cuda.manual_seed_all(2022)

# 2. 학습데이터 설정 X:input Y:output-> XOR (0=False, 1=True)
X = torch.FloatTensor([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]).to(device)

Y = torch.FloatTensor([
    [0], [1], [1], [0]
]).to(device)

# 3. Perceptron의 연결 구조 설정
model = nn.Sequential(
    nn.Linear(2, 1, bias=True),
    nn.Sigmoid()
).to(device)

# 4. 손실 함수와 학습 알고리즘 설정
loss_func = torch.nn.BCELoss().to(device)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

# 5. Perceptron 학습
for step in range(20001):
  optimizer.zero_grad()

  Y_hat = model(X)
  loss = loss_func(Y_hat, Y)
  loss.backward()
  optimizer.step()

  if step % 1000 == 0:
    print(step, loss.item())

  with torch.no_grad():
    model.eval()
    inputs = torch.FloatTensor([
        [0,0],
        [0,1],
        [1,0],
        [1,1]
    ]).to(device)
    outputs = model(inputs)

    print(outputs)
    print(outputs > 0.5)
```
